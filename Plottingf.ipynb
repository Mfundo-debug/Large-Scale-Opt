```python
from cvxopt import solvers, matrix
import numpy as np
import matplotlib.pyplot as plt
```

## Plot functions

```python
f = lambda x,y: 4.0*x + 2*y + x**2 - 4*x*y + y**2
x = np.linspace(-2, 6, 50)
y = np.linspace(-2, 6, 50)
x,y = np.meshgrid(x,y)
F = f(x,y)
cs = plt.contour(x,y,F,20)
plt.clabel(cs, inline =1, fontsize = 7)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
f = lambda x,y: 4.0*x + 2*y + x**2 - 4*x*y + y**2
x = np.linspace(-2, 6, 50)
y = np.linspace(-2, 6, 50)
x,y = np.meshgrid(x,y)
F = f(x,y)
fig = plt.figure(figsize=(8,6))
ax = fig.gca(projection = '3d')
ax.plot_surface(x,y,F, cmap = cm.coolwarm)
plt.show()
```

```python
f = lambda x,y: 4.0*x + 2*y + x**2 - 4*x*y + y**2
x = np.linspace(-2, 6, 50)
y = np.linspace(-2, 6, 50)
x,y = np.meshgrid(x,y)
F = f(x,y)
def plotter(E,A):
  fig = plt.figure(figsize=(8,6))
  ax = fig.gca(projection = '3d')
  ax.plot_surface(x,y,F, cmap = cm.coolwarm)
  ax.view_init(elev = E, azim= A)
  plt.show()

plotter(45,45)  
```
