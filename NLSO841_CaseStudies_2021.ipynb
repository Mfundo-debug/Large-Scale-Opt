# Using the Gradient Descent method to solve a simple equation.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define the objective function
def f(x):
    x1 = x[0]
    x2 = x[1]
    objFn = x1 + 2.0 * x2 + 4.0 * x1**2 - x1 * x2 + 2.0 * x2**2
    
    return objFn
```

```python
# define the objective gradient
def gradf(x):
    x1 = x[0]
    x2 = x[1]
    gradient = []
    gradient.append(1.0 + 8.0 * x1 - x2)
    gradient.append(2.0 - x1 + 4.0 * x2)
    
    return gradient
```

```python
# The Gradient Descnt Algorithm
itr = 10

# The learning rate
alpha = 0.15

# Initial solution
xStart = [-3.0, 2.0]

# Initialize the variables
xs = np.zeros((itr+1,2))
xs[0] = xStart

fx = []
#fx.append(f(xs[0]))

# The algorithm engine (GD)
for i in range(itr):
    xs[i+1] = xs[i] - np.dot(alpha,gradf(xs[i]))
    
    fx.append(f(xs[i+1]))
    
```

```python
# plot the graph
plt.plot(xs[:,0],xs[:,1],'g-o')

plt.xlabel('x1')
plt.ylabel('x2')

plt.show()
```

```python
line1 = np.arange(-4.0, 4.0, 0.1)
line2 = np.arange(-4.0, 4.0, 0.1)

x1Mesh, x2Mesh = np.meshgrid(line1,line2)
fMesh = x1Mesh + 2.0 * x2Mesh + 4.0 * x1Mesh**2 - x1Mesh * x2Mesh + 2.0 * x2Mesh**2

plt.figure(figsize = (8,6))

# specify the contour lines
lines = range(2,52,2)

# plot the contours
cs = plt.contour(x1Mesh,x2Mesh,fMesh,lines)

# label the contours
plt.clabel(cs, inline = 1, fontsize = 8)
plt.plot(xs[:,0],xs[:,1],'g-o')

plt.xlabel('x1')
plt.ylabel('x2')
plt.title('f(x) = x1 + 2x2 + 4x1^2 - x1x2 + 2x2^2')

plt.show()
```
