# Applying Gradient Descent method

```python
import numpy as np
import matplotlib.pyplot as plt
```

Define the objective function

```python
def f(x):
  x1= x[0]
  x2= x[1]
  ob_func = x1 + 2*x2 + 4*x1**2 -x1 *x2 + 2* x2**2
  return ob_func
```

```python
def gradf(x):
  x1= x[0]
  x2= x[1]
  gradient= []
  # input  partial derivatives for x1 and x2
  gradient.append(1 + 8*x1 - x2)
  gradient.append(2-x1 + 4*x2)

  return gradient
```

Gradient Descent algorithm

feel free to change iteration and learning rate

```python
iter = 10
#Learning rate
alpha = 0.15
#initial solution
x_str = [-2.0, 1.5]

#Initialize the variables
var_dim = np.zeros((iter+1, 2))
var_dim[0] = x_str

fx = []
fx.append(f(var_dim[0]))

# algorithm engine (gd)
for i in range(iter):
    var_dim[i+1] = var_dim[i] - np.dot(alpha, gradf(var_dim[i]))

    fx.append(f(var_dim[0]))

    
```

plot the graph

```python
plt.plot(var_dim[:,0], var_dim[:,1], 'g-o')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
line_1 = np.arange(-4,4,0.1)
line_2 = np.arange(-4,4,0.1)
x1Mesh, x2Mesh = np.meshgrid(line_1, line_2)
fMesh = x1Mesh + 2*x2Mesh + 4*x1Mesh**2 -x1Mesh *x2Mesh + 2* x2Mesh**2

plt.figure(figsize=(8,6))
#specify contour lines
lines = range(2,52,2)
#plot contours
cs = plt.contour(x1Mesh, x2Mesh, fMesh, lines)
#label the contour
plt.clabel(cs, inline=1, fontsize=10)
plt.plot(var_dim[:,0], var_dim[:,1], 'g-o')
plt.plot()
```
