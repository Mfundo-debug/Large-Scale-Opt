Question 3
minimizing $f(x)$ = $6 - 2x_1 + x_2 + 2x_1^2 +3x_1x_2 + x_2^2$ 

```python
#a)
def f(x):
  x1 = x[0]
  x2 = x[1]
  ob_f = 6 - 2*x1 + x2 + 2*x1**2 + 3*x1*x2  + x2**2
  return ob_f
```

```python
def gradf(x):
  x1 = x[0]
  x2 = x[1]
  gradient = []
  #input partial derivatives for x1 and x2
  gradient.append(-2 + 4*x1 +3*x2)
  gradient.append(1 +3*x2 + 3*x1)

  return gradient
```

gradient descent algorithm

```python
iter = 100
#learning rate
alpha = 0.05
#intial solution 
init_sol =[1.0, -2.0]

#initialize the variables
var = np.zeros((iter+1, 2))
var[0] = init_sol

fx = []
fx.append(f(var[0]))

#algorithm engine (gd)
for i in range(iter):
    var[i+1] = var[i] - np.dot(alpha, gradf(var[i]))

    fx.append(f(var[0]))
```

```python
plt.plot(var[:,0], var[:,1], 'g-o')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
#plot the graph
line_1 = np.arange(-4,4,0.1)
line_2 = np.arange(-4,4,0.1)
x1Mesh, x2Mesh = np.meshgrid(line_1, line_2)
fMesh = 6 - 2*x1Mesh + x2Mesh + 2*x1Mesh**2 +3*x1Mesh*x2Mesh + x2Mesh**2

plt.figure(figsize=(8,6))
#specify contour lines
lines = range(2,52,2)
#plot contours
cs = plt.contour(x1Mesh, x2Mesh, fMesh, lines)
#label the contour
plt.clabel(cs, inline=1, fontsize=10)
plt.plot(var[:,0], var[:,1], 'g-o')
plt.plot()
```

stochastic gradient descent

```python
from sklearn.metrics import mean_squared_error
```

```python
def GDM(X,y, step = 0.05, iteration = 100):
    
    intercept = 0.7
    slope = 0.5
    
    mse = []
    trackLog = []
    
    numRecord = len(X)
    
    for i in list(range(iteration)):
        f = y - (slope * X + intercept)
        
        slope = slope - step * (-2 * X.dot(f).sum() / numRecord)
        intercept = intercept - step * (-2 * f.sum() / numRecord)
        
        mse.append(mean_squared_error(y,(slope * X + intercept)))
        trackLog.append((slope,intercept))

    return slope,intercept,mse,trackLog
        
```

```python
np.random.seed(25)

X = np.array(sorted(list(range(5))*30)) + np.random.normal(size = 150, scale = 0.5)
y = np.array(sorted(list(range(5))*30)) + np.random.normal(size = 150, scale = 0.25)

slope,intercept,mse,trackLog = GDM(X,y)
f = slope * X + intercept
```
